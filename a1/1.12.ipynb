### b) ####

Let the affine set be denoted by $B$ and its associated subspace be denoted $A$. We need to shift the vector $x$ to project onto the subspace, then unshift the projection.

$$
\begin{aligned}
y^{(2)} &= proj_B(x - v_0) + v_0 \\
      &= \dfrac{<x - v_0, v>}{\norm{v}^2}v + v_0 \\
      &= \dfrac{<x - v_0, v>}{<v, v>}v + v_0 \\
      &= \dfrac{20}{29}v + v_0 \\
      &= \dfrac{20}{29}[-2, 5]^T + [-2, -2]^T \\
      &= [-3.38, -0.552]^T
\end{aligned}
$$

```matlab
%%file proj_cvx.m
%%capture
function [y , r ] = proj_cvx(x , v0 , v , nrm)%% x , v0 and v must be column vectors
    objtv = @(y) norm(x-y, nrm); %% objective is L_2 norm
    cvx_begin;
        variable y(2); %% 2 - d variable we are optimizing over
        variable t(1); %% real valued parameter that defines
        minimize(objtv(y)); %% defining the objective subject to
        v0 + t * v == y; %% the projection y must be in set A
    cvx_end;
    r = objtv(y); %% minimum value of the objective
end
```

```matlab
% Create inputs and optimize with cvx.
x = [3, 2]';
v = [-2 5]';
v0 = [-2 -4]';
[y1, r] = proj_cvx(x, v0, v, 1);
[y2, r] = proj_cvx(x, v0, v, 2);
[y3, r] = proj_cvx(x, v0, v, Inf);
```
